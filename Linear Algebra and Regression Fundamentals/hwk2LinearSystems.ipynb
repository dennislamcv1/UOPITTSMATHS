**Question 1**
Write code to solve $M\cdot \vec{x}=\vec{y}$ for $\vec{x}$. 

Let 
$
    M=\begin{bmatrix}
        1 & 2 & 3 \\ 
        5 & 5 & 2 \\
        7 & -3 & 11 \\
    \end{bmatrix}
$
and
$
    \vec{y} = \begin{bmatrix}
        88 \\
        208 \\
        135 \\
    \end{bmatrix}
$


```python
# Answer using NumPy
def solve_linear_system():
    M = np.array([
        [1, 2, 3],
        [5, 5, 2],
        [7, -3, 11]
    ])
    y = np.array([88, 208, 135])
    x = np.linalg.solve(M, y)
    return x

solve_linear_system()
```

```python
# Answer using SymPy
def solve_symbolic_system():
    M = sympy.Matrix([
        [1, 2, 3],
        [5, 5, 2],
        [7, -3, 11]
    ])
    y = sympy.Matrix([88, 208, 135])
    x = M.solve(y)
    return x

solve_symbolic_system()
```

**Question 2:**
You are a security consultant for an art gallery that has installed a new multi-sensor security system with 3 different types of sensors:
- Motion detectors (m)
- Heat sensors (h)
- Pressure plates (p)

The gallery conducted 3 test scenarios, and for each scenario, they recorded:
1. The total readings from each type of sensor across all rooms
2. The actual number of people that were in the gallery during each test

Your task is to determine how sensitive each type of sensor is to human presence.
In other words, you need to find out how to weight the output of each sensor in order to accurately predict the number of people in the room.

The test data with the actual number of people and each sensor's detected number of people:
- Test 1: 7 people  → m: 1, h: 2, p: 3
- Test 2: 8 person  → m: 5, h: 3, p: 0
- Test 3: 9 people  → m: 3, h: 2, p: 3

Your answer should return a vector a indicating the sensitivity of each sensor.

```python
import numpy as np
def calculate_sensitivity():
    # Rows: each test scenario [m, h, p]
    M = np.array([
        [1, 2, 3],
        [5, 3, 0],
        [3, 2, 3]
    ])
    # Actual number of people in each test
    people = np.array([7, 8, 9])
    # Solve M @ s = people for s (sensor sensitivities)
    s = np.linalg.solve(M, people)
    return s

calculate_sensitivity()
```
