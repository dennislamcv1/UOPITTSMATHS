**Question 1:** Use the following data set to and Numpy code to find the best fitting polynomial of the form
$$
    y = \beta_3x^3 + \beta_2 x^2 + \beta_1 x + \beta_0
$$

That is, solve for $\beta_3,\,\beta_2,\,\beta_1$ and $\beta_0$.

Represent your answer as an array variable named $beta$.

```python
data = pd.read_csv("./data/hw_data.csv")
data
```

```python
# Data for x and y defined outside the function
x = data['x'].values
y = data['y'].values

def calculate_matrix(x, y):
    # Create the matrix M using vstack and transpose it
    M = np.vstack([x**3, x**2, x, np.ones(len(x))]).T
    return M

# Call the function to create M
M = calculate_matrix(x, y)
```

```python
# HIDDEN TEST CELL
```

Since we don't have a square matrix, we must represent it as follows:
$$
(M^T \cdot M)\beta = M^Ty
$$

```python
def calculate_beta(M, y):
    # Use M.T to solve the normal equations: (M.T @ M) @ beta = M.T @ y
    G = M.T @ M
    rhs = M.T @ y
    beta = np.linalg.solve(G, rhs)
    return beta

beta = calculate_beta(M, y)
```

```python
# HIDDEN TEST CELL
```

Plot your results against the true `y` values.
Create a new variable `yhat` to represent your estimate.
Compare the results using [`np.linalg.norm`](https://numpy.org/doc/stable/reference/generated/numpy.linalg.norm.html).
Plot `yhat` and `y` with `^` and `o` respectively. Finally, provide a legend.

```python
def plot_estimates(M, beta, x, y):
    # Calculate predicted values (yhat)
    yhat = M @ beta
    # Print the norm of the residuals
    print('Residual norm:', np.linalg.norm(y - yhat))
    # Plot original and estimated
    plt.plot(x, y, 'o', label='y (actual)')
    plt.plot(x, yhat, '^', label='yhat (estimated)')
    plt.legend()

    # Do not edit or remove the code presented below.
    plt.title('Student Plot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.savefig('student_plot.png')
    plt.close()

plot_estimates(M, beta, x, y)
```

**Question 2:** Consider the data below. This data comes from the US Bureau of Labor Statistics. It includes the importance of a skills to all US occupations. There is also an "Annual Wage" column detailing the average annual wage for US workers of each occupation nationwide.

Use OLS regression to estimate the impact of each skill on an occupation's annual wage.

```python
data = pd.read_csv("./data/skillsData.csv")
data.head()
```

```python
import numpy as np
import pandas as pd

def calculate_sorted_beta(data):
    # Extract target variable and feature matrix
    y = data['Annual Wage'].values
    feature_cols = [c for c in data.columns if c != 'Annual Wage']
    M = data[feature_cols].values
    # Normal equations: (M.T @ M) @ beta = M.T @ y
    G = M.T @ M
    y2 = M.T @ y
    beta = np.linalg.solve(G, y2)
    # Return as sorted Pandas Series
    beta_series = pd.Series(beta, index=feature_cols).sort_values()
    return beta_series

calculate_sorted_beta(data)
```
