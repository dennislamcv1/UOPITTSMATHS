**Question 1:** The dataframe called "coinFlips" has two columns; the first represents the results for flipping Coin A and the second column represents the results of flipping Coin B.
In both columns, 1 represents a heads and 0 represents a tails.

Calcuate 
- $\hspace{.25cm}P(A=H)\hspace{.25cm}$ the probability that Coin A is heads, 
- $\hspace{.25cm}P(B=H)\hspace{.25cm}$ the probability the Coin B is heads, 
- and $\hspace{.25cm}P(A=H|B=H)\hspace{.25cm}$ the probability that Coin A is heads given that Coin B is heads (i.e., using the rows in the dataframe).

```python
coinFlips = pd.read_csv("./coinFlipData.csv",index_col=0)
coinFlips.shape
```

```python
def calculate_coin_probabilities():
    coinFlips = pd.read_csv("./coinFlipData.csv", index_col=0)
    
    # P(A = H): proportion of heads in Coin A
    pA_heads = coinFlips["Coin A"].mean()
    
    # P(B = H): proportion of heads in Coin B
    pB_heads = coinFlips["Coin B"].mean()
    
    # P(A = H | B = H): proportion of Coin A heads among rows where Coin B is heads
    pA_given_B_heads = coinFlips[coinFlips["Coin B"] == 1]["Coin A"].mean()
    
    return pA_heads, pB_heads, pA_given_B_heads

calculate_coin_probabilities()
```

**Question 2:** What is the p-value associated with the statistical test for the hypothesis that $\hspace{.25cm}P(A=H)>P(B=H)$? And, what is the p-value associated with the statistical test for the hypothesis that $\hspace{.25cm}P(A=H)>P(A=H|B=H)?$

```python
import scipy
from scipy import stats

def calculate_probabilities():
    coinFlips = pd.read_csv("./coinFlipData.csv", index_col=0)
    
    # Step 1: Calculate relevant probabilities
    pA_heads, pB_heads, pA_given_B_heads = calculate_coin_probabilities()
    
    n = len(coinFlips)
    kA = int(coinFlips["Coin A"].sum())
    
    # Step 2a: Test hypothesis P(A=H) > P(B=H)
    # Binomial test: observed A heads vs null probability = P(B=H), alternative = greater
    test1 = scipy.stats.binomtest(kA, n, pB_heads, alternative='greater')
    pval1 = test1.pvalue
    
    # Step 2b: Test hypothesis P(A=H) > P(A=H|B=H)
    # Equivalent to testing if P(A=H|B=H) < P(A=H)
    # Use binomial test on A heads within B=H rows vs null = P(A=H), alternative = less
    b_heads_mask = coinFlips["Coin B"] == 1
    nB = int(b_heads_mask.sum())
    kA_given_B = int(coinFlips[b_heads_mask]["Coin A"].sum())
    test2 = scipy.stats.binomtest(kA_given_B, nB, pA_heads, alternative='less')
    pval2 = test2.pvalue
    
    # Step 3: Return p-values
    return pval1, pval2

calculate_probabilities()
```

**Question 3:** The dataframe called "orderData" has two columns. The first column is sales data from Pizza Place 1 and the second columns is sales data from Pizza Place 2. What is the p-value for the hypothesis that the average sale per order is different between the two pizza places? 

Hint: Use a two-sided test.

```python
orderData = pd.read_csv("./orderData.csv",index_col=0)
orderData.shape
```

```python
def compare_sales():
    orderData = pd.read_csv("./orderData.csv", index_col=0)
    
    # Two-sided t-test comparing average sale per order between the two pizza places
    result = scipy.stats.ttest_ind(
        orderData["Pizza 1"].dropna(),
        orderData["Pizza 2"].dropna(),
        alternative='two-sided'
    )
    
    return result.pvalue

compare_sales()
```
