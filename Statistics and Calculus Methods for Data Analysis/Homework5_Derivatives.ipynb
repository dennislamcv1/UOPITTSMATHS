# Question 1
Use SymPy to find the partial derivative of $\hspace{.5cm}f(x,y) = x^2 \cdot y + y^3\hspace{.5cm}$ with respect to $x$ and the partial derivative with respect to $y$.

```python
# Using SymPy
x, y = sympy.symbols('x y')
f = x**2 * y + y**3
def calculate_derivative():   
    return sympy.diff(f, x)

calculate_derivative()
```

# Question 2
Using NumPy, numerically approximate the derivative of $\hspace{.5cm}f(x) = \cos(x) \hspace{.5cm}$ at $\hspace{.5cm}x = \frac{\pi}{3} \hspace{.5cm}$ using the central difference formula:
$$ f'(x) \approx \frac{f(x + h) - f(x )}{h} $$
Using  $\hspace{.5cm}h = 10^{-5}$

```python
#Using NumPy

def numerical_derivative():
    x = numpy.pi / 3
    h = 1e-5
    return (numpy.cos(x + h) - numpy.cos(x)) / h
    
numerical_derivative()
```

# Question 3
Use SymPy to find the exact derivative for $\hspace{.5cm}f(x) = \cos(x)\hspace{.5cm}$ for $x=\frac{\pi}{3}$

```python
# Use SymPy
def symbolic_derivative():
    x = sympy.symbols('x')
    f = sympy.cos(x)
    deriv = sympy.diff(f, x)
    return deriv.subs(x, sympy.pi / 3)
    

symbolic_derivative()
```
