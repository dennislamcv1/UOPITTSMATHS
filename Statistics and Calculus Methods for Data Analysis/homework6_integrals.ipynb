# Question 1
Use Sympy to solve $\hspace{.5cm}\displaystyle\int_2^{10}\left(\int_{-1}^5 4x^4+3y^3+2\,dx\right)dy$

```python
import sympy as sp

def Q1():
    # your code here
    x, y = sp.symbols('x y')
    expr = 4*x**4 + 3*y**3 + 2
    
    # Inner integral (with respect to x)
    inner = sp.integrate(expr, (x, -1, 5))
    
    # Outer integral (with respect to y)
    outer = sp.integrate(inner, (y, 2, 10))
    
    return outer


Q1()
```

# Question 2
Write code to calculate the Riemann sum to approximate $\hspace{.5cm}\displaystyle\int_2^4 5x^7+3x^2+2\,dx$

Use a partition with equally-sized bins of width $\hspace{.25cm}dx=.1$

In particular, use the partition $\hspace{.25cm}P=\{(2,2.1),\,(2.1,2.2),\dots,(3.9,4)\}$

Use left-side evaluation.

Numpy is allowed. Do not use Sympy.

```python
def Q2():
    # numpy.arange should be helpful for defining the particion
    # your code here
    # Left Riemann sum with dx = 0.1
    xs = np.arange(2, 4, 0.1)
    f = 5*xs**7 + 3*xs**2 + 2
    
    return np.sum(f * 0.1)


Q2()
```

# Question 3




Use SymPy to calculate the area under the Standard Normal Curve between $-0.4\sigma$ and $0.75\sigma$.


```python
# Sample Usage of sympy.lamdify
# x = sympy.symbols('x')
# expr = x**2 + 2*x + 1
# f = sympy.lambdify(x, expr)
# print(f(3))  # Outputs: 16

def Q3():
    # sympy.lambdify will be useful for evaluating symbolic functions for specific inputs
    # your code here
    x = sp.symbols('x')
    
    # Standard normal PDF
    pdf = (1 / sp.sqrt(2 * sp.pi)) * sp.exp(-x**2 / 2)
    
    # Integrate from -0.4 to 0.75 (σ cancels out)
    area = sp.integrate(pdf, (x, -0.4, 0.75))
    
    return sp.N(area)

Q3()
```
